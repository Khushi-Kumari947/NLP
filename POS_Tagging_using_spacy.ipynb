# POS(Parts of Speech Tagging)
In simple words, we can say that POS tagging is a tsk of lablling each word in a sentence with its appropiate part of speech.

Applications:

1.NER(Named Entity Recognintion)

2.Question Answering System

3.Word Sense Diambiguation

4.Chatbots

```python
import spacy
```

```python
# Importing english laguage library

nlp=spacy.load('en_core_web_sm')
```

```python
#Creating document-words can be extarcted using indexing
#automatically does pos tagging
doc=nlp(u"i'll google about facebook")
```

```python
doc.text
```

```python
doc[4]
```

```python
#corase grained pos tagging
doc[2].pos_
```

```python
#fine grined pos
doc[2].tag_
```

```python
spacy.explain('VB')
```

```python
for word in doc:
  print(word.text,'---->',word.pos_," ,",word.tag_,spacy.explain(word.tag_))
```

```python
doc2=nlp(u"i left the room")
for word in doc2:
  print(word.text,'---->',word.pos_," ,",word.tag_,spacy.explain(word.tag_))
```

```python
doc3=nlp(u"to the left of the room")
for word in doc3:
  print(word.text,'---->',word.pos_," ,",word.tag_,spacy.explain(word.tag_))
```
