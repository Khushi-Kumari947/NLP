```python
import os
import nltk
import gensim
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

# Download the specific resource for sentence splitting
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.corpus import stopwords

# Download the stop words list
nltk.download('stopwords')

# Create a set of English stop words (sets are faster for lookups than lists)
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_with_stopwords(cache_path):
    processed_sentences = []

    for filename in os.listdir(cache_path):
        if filename.endswith(".txt"):
            file_url = os.path.join(cache_path, filename)
            with open(file_url, 'r', encoding='utf-8', errors='ignore') as f:
                content = f.read()

                # 1. Split into sentences
                sentences = sent_tokenize(content)

                for s in sentences:
                    # 2. Tokenize and clean (Gensim)
                    tokens = simple_preprocess(s)

                    # 3. Filter out stop words
                    clean_tokens = [word for word in tokens if word not in stop_words]

                    if clean_tokens:
                        processed_sentences.append(clean_tokens)

    return processed_sentences

# Run the updated function
story = preprocess_with_stopwords(path)
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=story ,
    vector_size=150,  # Slightly higher for better character nuances
    window=10,         # GoT has long descriptive sentences; a larger window helps
    min_count=2,      # Ignore rare typos
    workers=4
)

# Access a vector
# print(model.wv['daenerys'])
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```
