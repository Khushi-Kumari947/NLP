```python
#remove tags
#change to lowercase
#remove stopwords
```

```python
import re
def remove_tags(raw_text):
  cleaned_text=re.sub(re.compile('<.*?>'),'',raw_text)
  return cleaned_text
```

```python
new_df['review']=new_df['review'].apply(remove_tags)
```

```python
new_df['review'][189]
```

```python
!pip install nltk
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
sw_list=stopwords.words('english')
new_df['review']=new_df['review'].apply(lambda x: [item for item in x.split() if item not in sw_list])
```

```python
new_df['review'].sample(3)
```

```python
X=new_df.iloc[:,0:1]
```

```python
X.shape
```

```python
y=new_df['sentiment']
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

def dummy(doc):
    return doc

cv = CountVectorizer(tokenizer=dummy, preprocessor=dummy, lowercase=False)
X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow=cv.transform(X_test['review']).toarray()
```

```python
print(cv.get_feature_names_out()[:10])
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train_bow,y_train)
```

```python
y_pred=gnb.predict(X_test_bow)
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(X_train_bow,y_train)
y_pred=rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```
